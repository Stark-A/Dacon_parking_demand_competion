```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
train = pd.read_csv("../Data/parking_demand/train_df_errno.csv")
test = pd.read_csv("../Data/parking_demand/test_df.csv")
sub = pd.read_csv("../Data/parking_demand/sample_submission.csv")
age = pd.read_csv("../Data/parking_demand/age_gender_info.csv")
train.shape, test.shape, sub.shape, age.shape
```

```python
gangwon = train.loc[train['지역']== '강원도']
g_d = gangwon.drop('10분내지하철수',axis=1)
```

```python
g_d.head()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=g_d, y='등록차량수')
```

```python
g_d.shape
```

```python
g1 = g_d.dropna(how='all', axis=1).dropna(how='all', axis=0)
```

```python
g_c = g1.corr()
```

```python
mask = np.triu(np.ones_like(g_c))
```

```python
plt.figure(figsize=(12, 5))
sns.heatmap(g_c, annot=True, fmt='.2f', mask=mask, cmap='seismic', vmax=1, vmin=-1)
```

```python
t1 = t_d.dropna(how='all', axis=1).dropna(how='all', axis=0)
```
